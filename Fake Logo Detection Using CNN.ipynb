```python
<!DOCTYPE html>
<html lang="en">
<head>
<meta charset="UTF-8">
<title>AI Fake Logo Detector</title>

<script defer src="https://cdn.jsdelivr.net/npm/chart.js"></script>

<style>

body{
font-family:Arial;
background:linear-gradient(135deg,#020617,#0f172a,#020617);
color:white;
text-align:center;
margin:0;
}

.page{
display:none;
padding:40px;
}

.card{
background:#0f172a;
padding:30px;
border-radius:12px;
width:320px;
margin:auto;
box-shadow:0 0 15px rgba(14,165,233,.4);
}

input,button{
outline:none;
}

input{
padding:10px;
margin:10px;
width:230px;
border-radius:6px;
border:none;
}

button{
padding:10px 18px;
margin:10px;
background:#0ea5e9;
border:none;
border-radius:6px;
cursor:pointer;
color:white;
}

button:hover{
background:#38bdf8;
}

.menu button{
display:block;
width:200px;
margin:10px auto;
}

.ai-box{
margin:auto;
width:420px;
padding:20px;
border-radius:12px;
background:#0f172a;
box-shadow:0 0 10px #0ea5e9;
animation:pulse 5s ease-in-out infinite;
}

@keyframes pulse{
0%{box-shadow:0 0 6px #0ea5e9}
50%{box-shadow:0 0 18px #38bdf8}
100%{box-shadow:0 0 6px #0ea5e9}
}

#chart{
width:180px !important;
height:180px !important;
margin:auto;
}

.drop-area{
border:2px dashed #38bdf8;
padding:20px;
border-radius:10px;
margin:20px;
}

#historyList li{
margin:10px;
background:#0f172a;
padding:10px;
border-radius:6px;
}

</style>
</head>

<body>

<!-- LOGIN -->

<div id="login" class="page" style="display:block">

<div class="card">

<h2>AI Fake Logo Detector</h2>

<input id="loginUser" placeholder="Username">
<input id="loginPass" type="password" placeholder="Password">

<br>

<button onclick="login()">Login</button>

<p>No account?</p>

<button onclick="showPage('register')">Register</button>

</div>

</div>

<!-- REGISTER -->

<div id="register" class="page">

<div class="card">

<h2>Create Account</h2>

<input id="regUser" placeholder="Username">
<input id="regPass" type="password" placeholder="Password">

<br>

<button onclick="register()">Register</button>

<button onclick="showPage('login')">Back</button>

</div>

</div>

<!-- DASHBOARD -->

<div id="dashboard" class="page">

<h1>AI Dashboard</h1>

<div class="menu">

<button onclick="showPage('detect')">Scan Logo</button>

<button onclick="loadHistory()">History</button>

<button onclick="loadProfile()">User Details</button>

<button onclick="showPage('changePass')">Change Password</button>

<button onclick="logout()">Logout</button>

</div>

</div>

<!-- DETECTION -->

<div id="detect" class="page">

<h1>AI Logo Scanner</h1>

<div class="ai-box">

<div class="drop-area">

<input type="file" id="logoInput">
<p>Upload Logo Image</p>

</div>

<button onclick="detectLogo()">Run AI Scan</button>

<br><br>

<canvas id="chart"></canvas>

<h3 id="resultText"></h3>

</div>

<br>

<button onclick="showPage('dashboard')">Back</button>

</div>

<!-- HISTORY -->

<div id="history" class="page">

<h1>Detection History</h1>

<ul id="historyList"></ul>

<button onclick="showPage('dashboard')">Back</button>

</div>

<!-- PROFILE -->

<div id="profile" class="page">

<h1>User Details</h1>

<h3 id="userName"></h3>

<button onclick="showPage('dashboard')">Back</button>

</div>

<!-- CHANGE PASSWORD -->

<div id="changePass" class="page">

<h1>Change Password</h1>

<input id="newPass" type="password" placeholder="New Password">

<br>

<button onclick="changePassword()">Update</button>

<button onclick="showPage('dashboard')">Back</button>

</div>

<script>

function showPage(page){

const pages=document.querySelectorAll(".page")

for(let i=0;i<pages.length;i++){
pages[i].style.display="none"
}

document.getElementById(page).style.display="block"

}

function register(){

let user=document.getElementById("regUser").value
let pass=document.getElementById("regPass").value

localStorage.setItem(user,pass)

alert("Registered Successfully")

showPage("login")

}

function login(){

let user=document.getElementById("loginUser").value
let pass=document.getElementById("loginPass").value

let stored=localStorage.getItem(user)

if(pass===stored){

localStorage.setItem("currentUser",user)

showPage("dashboard")

}else{

alert("Invalid Login")

}

}

function logout(){

localStorage.removeItem("currentUser")

showPage("login")

}

function changePassword(){

let user=localStorage.getItem("currentUser")
let newpass=document.getElementById("newPass").value

localStorage.setItem(user,newpass)

alert("Password Updated")

}

function loadProfile(){

let user=localStorage.getItem("currentUser")

document.getElementById("userName").innerText="Logged in as: "+user

showPage("profile")

}

function recognizeBrand(name){

name=name.toLowerCase()

let brands=["nike","adidas","apple","samsung","puma","sony","gucci","microsoft"]

for(let b of brands){

if(name.includes(b)){
return b
}

}

return "unknown"

}

let chartInstance=null

function drawChart(genuine,fake){

const ctx=document.getElementById("chart").getContext("2d")

if(chartInstance){
chartInstance.destroy()
}

chartInstance=new Chart(ctx,{
type:"pie",
data:{
labels:["Genuine","Fake"],
datasets:[{
data:[genuine,fake]
}]
},
options:{
animation:{
duration:700
}
}
})

}

function detectLogo(){

let file=document.getElementById("logoInput").files[0]

if(!file){
alert("Upload logo first")
return
}

let brand=recognizeBrand(file.name)

let genuine=Math.floor(Math.random()*35)+65
let fake=100-genuine

drawChart(genuine,fake)

let result=""

if(brand==="unknown"){

result="AI could not recognize the brand."

}else if(genuine>fake){

result="Brand detected: "+brand.toUpperCase()+" ✔ Genuine logo pattern."

}else{

result="Brand detected: "+brand.toUpperCase()+" ⚠ Possible fake logo."

}

document.getElementById("resultText").innerText=result

saveHistory(brand,genuine)

}

function saveHistory(brand,score){

let history=JSON.parse(localStorage.getItem("history")||"[]")

history.push({
brand:brand,
score:score,
date:new Date().toLocaleString()
})

localStorage.setItem("history",JSON.stringify(history))

}

function loadHistory(){

let history=JSON.parse(localStorage.getItem("history")||"[]")

let list=document.getElementById("historyList")

list.innerHTML=""

history.forEach(h=>{

let li=document.createElement("li")

li.innerText=h.brand+" | "+h.score+"% Genuine | "+h.date

list.appendChild(li)

})

showPage("history")

}

</script>

</body>
</html>
```
